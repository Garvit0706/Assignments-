```python
Q3: Compare Pearson and Spearman correlations between exercise hours and BMI
Data (hypothetical example for illustration):

Exercise Hours: 
ùëã= [data]
BMI:ùëå=[data]

import numpy as np
from scipy.stats import spearmanr

# Hypothetical data for 50 participants
np.random.seed(0)
exercise_hours = np.random.randint(1, 10, 50)
bmi = np.random.randint(18, 30, 50)

# Pearson correlation coefficient
pearson_corr = np.corrcoef(exercise_hours, bmi)[0, 1]

# Spearman's rank correlation
spearman_corr, _ = spearmanr(exercise_hours, bmi)

pearson_corr, spearman_corr
```

```python
Q5: Calculating covariance matrix
Given the hypothetical data for age and soft drink preference (e.g., encoded numerically).

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Example data
data = pd.DataFrame({
    'Age': [23, 45, 31, 34],
    'SoftDrinkPreference': ['BrandA', 'BrandB', 'BrandA', 'BrandC']
})

# Label encoding
label_encoder = LabelEncoder()
data['SoftDrinkPreference_Encoded'] = label_encoder.fit_transform(data['SoftDrinkPreference'])

# Covariance matrix
cov_matrix = data[['Age', 'SoftDrinkPreference_Encoded']].cov()
print(cov_matrix)
```
