```python
2.
import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Calculate the standard error
    standard_error = sample_std_dev / math.sqrt(sample_size)
    return sample_mean, standard_error

# Example usage
sample_mean = 500
sample_std_dev = 50
sample_size = 50
mean, std_error = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print("Sample Mean:", mean)
print("Standard Error:", std_error)
```

```python
Q3: What is Hypothesis Testing? Why is it used? Importance of Hypothesis Testing
Hypothesis Testing:

A statistical method that uses sample data to evaluate a hypothesis about a population parameter.
Helps determine whether there is enough evidence to reject a null hypothesis in favor of an alternative hypothesis.
Importance:

Provides a formal process for making inferences about population parameters based on sample data.
Helps in making data-driven decisions in various fields such as medicine, economics, and engineering.
```

```python
Q8: Define p-value and Explain Its Significance in Hypothesis Testing
p-value:

The probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true.
It quantifies the evidence against the null hypothesis.
Significance:

A small p-value (typically ‚â§ 0.05) indicates strong evidence against the null hypothesis, leading to its rejection.
A large p-value (> 0.05) indicates weak evidence against the null hypothesis, so it is not rejected.
```

```python
Q14: Hypothesis Test for Blood Pressure Decrease
Given:

Sample mean (
ùë•
Àâ
x
Àâ
 ) = 8 mmHg
Sample standard deviation (
ùë†
s) = 3 mmHg
Sample size (
ùëõ
n) = 100
Hypothesized mean decrease (
ùúá
0
Œº 
0
‚Äã
 ) = 10 mmHg
Significance level (
ùõº
Œ±) = 0.05
Hypotheses:

ùêª
0
:
ùúá
=
10
H 
0
‚Äã
 :Œº=10
ùêª
1
:
ùúá
‚â†
10
H 
1
‚Äã
 :Œº
ÓÄ†
=10
Using the formula for the t-statistic:

ùë°
=
ùë•
Àâ
‚àí
ùúá
0
ùë†
/
ùëõ
t= 
s/ 
n
‚Äã
 
x
Àâ
 ‚àíŒº 
0
‚Äã
 
‚Äã
import scipy.stats as stats

# Given data
sample_mean = 8
hypothesized_mean = 10
sample_std_dev = 3
sample_size = 100
alpha = 0.05

# t-statistic
t_stat = (sample_mean - hypothesized_mean) / (sample_std_dev / np.sqrt(sample_size))

# p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=sample_size - 1))

print("t-statistic:", t_stat)
print("p-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
