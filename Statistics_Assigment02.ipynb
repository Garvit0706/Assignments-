```python
Q1. What are the three measures of central tendency?
The three measures of central tendency are:

Mean: The average of all the data points.
Median: The middle value when the data points are arranged in ascending order.
Mode: The most frequently occurring value in the dataset.
```

Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?
Mean: Calculated as the sum of all values divided by the number of values. It is used to find the average and is sensitive to outliers.
Median: The middle value in a sorted dataset. It is used when the dataset has outliers or is skewed, as it provides a better central value without being affected by extreme values.
Mode: The most frequently occurring value. It is useful in categorical data where we want to know the most common category

Q8. What do you understand about skewness in data?
Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. If the skewness is negative, the data is left-skewed (tail on the left). If it is positive, the data is right-skewed (tail on the right).

```python
Q13. How is covariance different from correlation?
Covariance:

Measures the direction of the linear relationship between variables.
Units are the product of the units of the two variables.
Correlation:

Measures both the strength and direction of the linear relationship.
Unitless, standardized value between -1 and 1.
```
