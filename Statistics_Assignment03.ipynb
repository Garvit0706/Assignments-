```python
3.
import math

def normal_pdf(x, mean, std_dev):
    """Calculate the Probability Density Function of a Normal Distribution."""
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    return coefficient * exponent

# Example usage
mean = 0
std_dev = 1
x = 0
pdf_value = normal_pdf(x, mean, std_dev)
print(f"PDF value at x={x} for N({mean}, {std_dev}^2) is {pdf_value}")
```

Q4. Properties of Binomial Distribution and Examples
Properties:

The experiment consists of 
𝑛
n independent trials.
Each trial has only two possible outcomes: success or failure.
The probability of success, 
𝑝
p, remains constant for each trial.
The random variable 
𝑋
X represents the number of successes in 
𝑛
n trials.
Examples:

Tossing a coin 10 times and counting the number of heads.
Conducting a survey where respondents can either favor or oppose a policy, and counting the number of favorable responses out of 100 respondents.

```python
5.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # number of trials
p = 0.4  # probability of success
size = 1000  # sample size

# Generate random sample
binomial_sample = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(binomial_sample, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
6.
import math

def poisson_cdf(k, mean):
    """Calculate the Cumulative Distribution Function of a Poisson Distribution."""
    cumulative_prob = 0
    for i in range(k + 1):
        cumulative_prob += (math.exp(-mean) * mean ** i) / math.factorial(i)
    return cumulative_prob

# Example usage
mean = 5
k = 3
cdf_value = poisson_cdf(k, mean)
print(f"CDF value at k={k} for Poisson({mean}) is {cdf_value}")
```

```python
Q7. Differences between Binomial Distribution and Poisson Distribution

Binomial Distribution:
Describes the number of successes in a fixed number of independent trials.
Each trial has two possible outcomes: success or failure.
Probability of success is constant for each trial.
Parameters: 
𝑛
n (number of trials) and 
𝑝
p (probability of success).

Poisson Distribution:
Describes the number of events occurring in a fixed interval of time or space.
Events occur independently and at a constant average rate.
Suitable for rare events.
Parameter: 
𝜆
λ (mean number of events in the interval).
```

```python
8.
import numpy as np

# Parameters
mean = 5
size = 1000

# Generate random sample
poisson_sample = np.random.poisson(mean, size)

# Calculate sample mean and variance
sample_mean = np.mean(poisson_sample)
sample_variance = np.var(poisson_sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

Q10. In Normal Distribution, Where Does the Least Frequent Data Appear?
In a normal distribution, the least frequent data appears in the tails, farthest from the mean. These regions correspond to values that are many standard deviations away from the mean.
