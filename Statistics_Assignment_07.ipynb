```python
1.
import scipy.stats as stats
import numpy as np

# Given data
mean = 50
std_dev = 5
sample_size = 30  # Assume a sample size of 30 for this example

# Calculate the standard error
standard_error = std_dev / np.sqrt(sample_size)

# Calculate the 95% confidence interval
confidence_level = 0.95
degrees_freedom = sample_size - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, mean, standard_error)

print("95% confidence interval:", confidence_interval)
```

```python
2.
import scipy.stats as stats

# Observed frequencies
observed = [22, 18, 24, 9, 10, 17]  # Example observed frequencies for each color

# Expected frequencies
total = sum(observed)
expected = [0.20 * total, 0.20 * total, 0.20 * total, 0.10 * total, 0.10 * total, 0.20 * total]

# Perform chi-square goodness of fit test
chi_square_stat, p_value = stats.chisquare(observed, expected)

print("Chi-square statistic:", chi_square_stat)
print("P-value:", p_value)
```

```python
3.
import numpy as np
import scipy.stats as stats

# Example contingency table data
#              | Yes | No |
# Smoker       | 30  | 70 |
# Non-Smoker   | 10  | 90 |

observed = np.array([[30, 70], [10, 90]])

# Perform chi-square test of independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(observed)

print("Chi-square statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

```python
6.
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Chi-square distribution parameters
df = 10
x = np.linspace(0, 30, 500)

# Plot chi-square distribution
plt.plot(x, stats.chi2.pdf(x, df), 'b-', label=f'Chi-square df={df}')
plt.fill_between(x, stats.chi2.pdf(x, df), where=(x <= 15), color='gray', alpha=0.5)

# Label axes and title
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-square value')
plt.ylabel('Probability density')
plt.legend()
plt.show()
```

```python
10.
import numpy as np
import scipy.stats as stats

# Create the observed frequency table
observed = np.array([[150, 200, 150], [200, 150, 150]])

# Perform chi-square test for independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(observed)

print("Chi-square statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

# Interpretation
significance_level = 0.01
if p_val < significance_level:
    print("Reject the null hypothesis. There is a significant association between chocolate preference and country of origin.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between chocolate preference and country of origin.")
```

```python
11.
import scipy.stats as stats
import numpy as np

# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70
alpha = 0.05

# Calculate the t-statistic
t_stat = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Degrees of freedom
dof = sample_size - 1

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), dof))

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. The population mean is not significantly different from 70.")
```
