```python
Q1. What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
Difference between t-test and z-test:

Population Variance:

t-test: Used when the population variance is unknown.
z-test: Used when the population variance is known.
Sample Size:

t-test: Typically used when the sample size is small (n < 30).
z-test: Typically used when the sample size is large (n ≥ 30).
Distribution:

t-test: Based on the t-distribution.
z-test: Based on the normal distribution (z-distribution).

# Example code for t-test using scipy
from scipy import stats

sample_heights = [150, 160, 170, 180, 190]  # Example data
t_stat, p_value = stats.ttest_1samp(sample_heights, 165)
print(t_stat, p_value)


# Example code for z-test
import numpy as np
from scipy import stats

sample_mean = 102
population_mean = 100
population_std = 15
n = 100

z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_stat)))
print(z_stat, p_value)
```

```python
Q2. Differentiate between one-tailed and two-tailed tests.
One-tailed test:

Tests for a deviation in only one direction (either greater than or less than).
Hypotheses:
H0: μ = μ0
H1: μ > μ0 (right-tailed) or H1: μ < μ0 (left-tailed)
Two-tailed test:

Tests for deviations in both directions (both greater than and less than).
Hypotheses:
H0: μ = μ0
H1: μ ≠ μ0
Example:

One-tailed test: Testing if a new drug increases the mean lifespan of patients more than the current drug.

Two-tailed test: Testing if a new teaching method affects the average test scores (it could be higher or lower than the current method).
```

```python
3. Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
Type 1 error (α): Rejecting the null hypothesis when it is actually true.

Example: A court convicting an innocent person (false positive).
Type 2 error (β): Failing to reject the null hypothesis when it is actually false.

Example: A court acquitting a guilty person (false negative).
```

Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.
Sample problem:

A factory has two machines: Machine A produces 60% of the items, and Machine B produces 40% of the items.
Machine A has a defect rate of 1%, and Machine B has a defect rate of 2%.
An item is selected at random and found to be defective. What is the probability it was produced by Machine A?
Solution:

Prior probabilities: P(A) = 0.60, P(B) = 0.40
Likelihoods: P(D|A) = 0.01, P(D|B) = 0.02
Total probability of defect: P(D) = P(D|A)P(A) + P(D|B)P(B)
